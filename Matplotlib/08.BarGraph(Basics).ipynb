# Bar Graph (Basics)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # 맑은 고딕
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
labels = ['강백호', '서태웅', '정대만'] # name
values = [190, 187, 184] # height

plt.bar(labels, values)
```

```python
labels = ['강백호', '서태웅', '정대만'] # name
values = [190, 187, 184] # height

plt.bar(labels, values, color='r')
```

```python
labels = ['강백호', '서태웅', '정대만'] # name
values = [190, 187, 184] # height
colors=['r', 'g', 'b']

plt.bar(labels, values, color=colors, alpha=0.7)
```

```python
labels = ['강백호', '서태웅', '정대만'] # name
values = [190, 187, 184] # height

plt.bar(labels, values)
plt.ylim(175, 195) # start and end of y-axis
```

```python
plt.bar(labels, values, width=0.5) # narrower bar graph (width: 0~1)
```

```python
plt.bar(labels, values, width=0.3)
plt.xticks(rotation=45) # tilt names in x-axis
plt.yticks(rotation=45) # tilt values in y-axis
```

```python
labels = ['강백호', '서태웅', '정대만'] # name
values = [190, 187, 184] # height
ticks = ['1번학생', '2번학생', '3번학생']

plt.bar(labels, values)
plt.xticks(labels, ticks, rotation=45)
```
