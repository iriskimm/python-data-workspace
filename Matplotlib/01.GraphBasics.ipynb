# 01. Matplotlib
A library used to visualize data through various kinds of graph

```python
import matplotlib.pyplot as plt
```

## 1. Graph Basics

```python
x = [1, 2, 3]
y = [2, 4, 8]
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.show() # excludes the line that shows location
```

## Setting the Title

```python
plt.plot(x, y)
plt.title('Line Graph')
```

## Font Configuration

```python
# Configuration for Korean font
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # 맑은 고딕
```

```python
# Configure font size
matplotlib.rcParams['font.size'] = 15
```

```python
# allow to use minus signs
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# See what kinds of font are available
import matplotlib.font_manager as fm
fm.fontManager.ttflist
[f.name for f in fm.fontManager.ttflist] # only print names for fonts 
```

```python
plt.plot(x, y)
plt.title('그래프') # now we can write titles in korean
```
