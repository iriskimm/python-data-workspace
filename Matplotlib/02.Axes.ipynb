# 02. Axes

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # 맑은 고딕
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
plt.title('Line Graph', fontdict={'family':'Times New Roman', 'size':20}) # set font individually
```

```python
plt.plot(x, y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

```python
plt.plot(x, y)
plt.xlabel('X axis', color='purple')
plt.ylabel('Y axis', color='#00aa00')
```

```python
plt.plot(x, y)
plt.xlabel('X axis', color='purple', loc='right') # left, center, right
plt.ylabel('Y axis', color='#00aa00', loc='top') # top, center, bottom
```

```python
plt.plot(x, y)
plt.xticks([1, 2, 3])
plt.yticks([3, 6, 9, 12])
plt.show()
```
