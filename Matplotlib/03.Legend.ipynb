# 03. Legend

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # 맑은 고딕
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, label='data 1')
plt.legend()
```

```python
plt.plot(x, y, label='data 1')
plt.legend(loc='upper right') # put legend on upper right (it can be lower right, center, center right, best, etc.)
```

```python
plt.plot(x, y, label='data 1')
plt.legend(loc='best') # it's best by default
```

```python
plt.plot(x, y, label='legend')
plt.legend(loc=(0.5, 0.5)) # x-axis, y-axis (between 0~1)
```
