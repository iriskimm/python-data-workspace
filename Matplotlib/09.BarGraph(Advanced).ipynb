# 9. Bar Graph (Advanced)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # 맑은 고딕
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
labels = ['강백호', '서태웅', '정대만'] # name
values = [190, 187, 184] # height

plt.barh(labels, values) # horizontal bar graph
plt.xlim(175, 195)
```

```python
bar = plt.bar(labels, values)
bar[0].set_hatch('/') # fills in the bar with ///
bar[1].set_hatch('x') # xxxx
bar[2].set_hatch('..') # ...
```

```python
bar = plt.bar(labels, values)
plt.ylim(175, 195)

for idx, rect in enumerate(bar): 
    plt.text(idx, rect.get_height() + 0.5, values[idx], ha='center', color='skyblue')
```
