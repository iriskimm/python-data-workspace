# 04. Style

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # 맑은 고딕
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, linewidth=5)
```

***
## Marker

```python
plt.plot(x, y, marker='o')
```

```python
plt.plot(x, y, marker='o', linestyle='None')
```

```python
plt.plot(x, y, marker='v')
```

```python
plt.plot(x, y, marker='v', markersize=10)
```

```python
plt.plot(x, y, marker='X', markersize=10)
```

```python
plt.plot(x, y, marker='o', markersize=10, markeredgecolor='red')
```

```python
plt.plot(x, y, marker='o', markersize=10, markeredgecolor='red', markerfacecolor='yellow')
```

***
## Line Style

```python
plt.plot(x, y, linestyle=':')
```

```python
plt.plot(x, y, linestyle='--')
```

```python
plt.plot(x, y, linestyle='-.')
```

```python
plt.plot(x, y, linestyle='-')
```

***
## Color

```python
plt.plot(x, y, color='pink')
```

```python
plt.plot(x, y, color='#ff0000')
```

```python
plt.plot(x, y, color='b')
```

```python
plt.plot(x, y, color='g')
```

***
## Format (everything combined)

```python
plt.plot(x, y, 'ro--') # color: r, marker: 'o', linestyle='--'
```

```python
plt.plot(x, y, 'bv:')
```

```python
plt.plot(x, y, 'go') # green, 'o', linestyle='None'
```

***
## Abbreviations

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, mec='blue', ls='-.') # mfc: markerfacecolor, ms: markersize, mec: markeredgecolor, ls: linestyle
```

***
## Opacity

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, alpha=0.3) # aplha implies opacity (0~1)
```

***
## Graph Size

```python
plt.figure(figsize=(10, 5)) # width, height
plt.plot(x, y)
```

```python
plt.figure(figsize=(5, 10))
plt.plot(x, y)
```

```python
plt.figure(figsize=(10, 5), dpi=150) # dots per inch - 해상도(확대)
plt.plot(x, y)
```

***
## Background Color

```python
plt.figure(facecolor='yellow')
plt.plot(x, y)
```

```python
plt.figure(facecolor='#a1c3ff')
plt.plot(x, y)
```
