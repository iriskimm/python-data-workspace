# 16. Multiple Graphs

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # 맑은 고딕
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10)) # 2 * 2 에 해당하는 plot 들을 생성
fig.suptitle('Multiple Graphs')

# First Graph
axs[0, 0].bar(df['이름'], df['국어'], label='국어 점수')
axs[0, 0].set_title('First Graph')
axs[0, 0].legend()
axs[0, 0].set(xlabel='이름', ylabel='점수')
axs[0, 0].set_facecolor('lightyellow')
axs[0, 0].grid(linestyle='--', linewidth=0.5)
axs[0, 0].tick_params(axis='x', rotation=45)

# Second Graph
axs[0, 1].plot(df['이름'], df['수학'], label='수학')
axs[0, 1].plot(df['이름'], df['영어'], label='영어')
axs[0, 1].legend()

# Third Graph
axs[1, 0].barh(df['이름'], df['키'])
axs[1, 0].set_xlim(150, 210)

# Fourth Graph
axs[1, 1].plot(df['이름'], df['사회'], color='green', alpha=0.5)
```
