# 14. Pie Chart (Advanced)

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # 맑은 고딕
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']
colors = ['#ffadad', '#ffd6a5', '#fdffb6', '#caffbf', '#9bf6ff', '#a0c4ff']
explode = [0.03] * 6
```

```python
plt.pie(values, labels=labels, autopct='%.1f%%', colors=colors, explode=explode, startangle=90, counterclock=False)
plt.show()
```

```python
wedgeprops = {'width': 0.6}
plt.pie(values, labels=labels, autopct='%.1f%%', colors=colors, explode=explode, startangle=90, counterclock=False, wedgeprops=wedgeprops)
plt.show()
```

```python
wedgeprops = {'width': 0.6, 'edgecolor':'w', 'linewidth':2}
plt.pie(values, labels=labels, autopct='%.1f%%', colors=colors, startangle=90, counterclock=False, wedgeprops=wedgeprops)
plt.show()
```

```python
def custom_autopct(pct): 
#     return ('%.1f%%' % pct) if pct >= 10 else ''
#     return '{:.1f}%'.format(pct) if pct >= 10 else ''
    return '{:.0f}%'.format(pct) if pct >= 10 else ''

plt.pie(values, labels=labels, autopct=custom_autopct, colors=colors, startangle=90, counterclock=False, wedgeprops=wedgeprops, pctdistance=0.7)
plt.show()
```
