# Edit Data

```python
import pandas as pd
```

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

## Edit Data from a Column

```python
df['학교'].replace({'북산고':'상북고', '능남고':'무슨고'}, inplace=True) # pass arguments as a dictionary
df
```

```python
df['SW특기'] = df['SW특기'].str.lower()
df
```

```python
df['SW특기'] = df['SW특기'].str.upper()
df
```

```python
df['학교'] = df['학교'] + '등학교'
df
```

## Add a Column

```python
df['총합'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
df
```

```python
df['평균'] = df['총합']/5
df
```

```python
df['결과'] = 'Fail' # Fail by default
df.loc[df['총합'] > 400, '결과'] = 'Pass' # choose row and column, set it to pass
df
```

## Delete a Column

```python
df.drop(columns='총합')
```

```python
df.drop(columns=['국어', '영어', '수학'])
```

```python
df
```

```python
df.drop(columns='총합', inplace=True)
```
