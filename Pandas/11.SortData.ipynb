# Sort Data

```python
import pandas as pd
```

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.sort_values('키') # sort students by their height in ascending order
```

```python
df.sort_values('키', ascending=False) # in descending order
```

```python
df.sort_values(['수학', '영어'], ascending=False) # sort by math, and also english if math values are same
```

```python
df.sort_values(['수학', '키'], ascending=[False, True]) # sort math in descending order, and heigh in ascending order
```

```python
df['키'].sort_values()
```

```python
df.sort_index()
```

```python
df.sort_index(ascending=False)
```
