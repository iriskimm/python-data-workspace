# 3. Index
An address to access the data

```python
import pandas as pd
```

```python
data = {
    '이름' : ['채치수', '정대만', '송태섭', '서태웅', '강백호', '변덕규', '황태산', '윤대협'],
    '학교' : ['북산고', '북산고', '북산고', '북산고', '북산고', '능남고', '능남고', '능남고'],
    '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}
```

```python
df = pd.DataFrame(data, index=['1', '2', '3', '4', '5', '6', '7', '8'])
df
```

```python
df.index # prints out the index
```

## Setting index property(name)

```python
df.index.name = '지원번호'
df
```

## Reset(remove) index

```python
df.reset_index()
```

```python
df.reset_index(drop=True) # also remove the added index name (지원번호)
```

```python
df.reset_index(drop=True, inplace=True) # need to manually apply to save changes to the data
```

```python
df
```

## Choose Index from one of the Columns

```python
df.set_index('이름', inplace=True) # raises an error if index was already set to be '이름'
```
